***The keys() method will return a list of all the keys in the dictionary.***

```python
car_dict={"brand":"BMW","model":"iX","year":1964}
car_dict.keys()
```

```python
#The values() method will return a list of all the values in the dictionary.
car_dict={"brand":"BMW","model":"iX","year":1964}
car_dict.values()
```

***Update Method***

***Removing Dictionary items for Pop() Method***

```python
#The pop() method removes the item with the specified key name:
d3={"name":"A",'year':2023,'Pass':True,'Marks':[50,60,70]}
d3.pop('Marks')
```

```python
d3
```

***Popitems() Method***

```python
#The popitem() method removes the last inserted item
d3 = {'name': 'A','year': 2024,'Pass': True,'Marks': [50, 60, 70],'phone_no': 94566,'email': 'asdf@gmail.com'}
d3.popitem()
```

```python
d3
```

***Del Method***

```python
#The del keyword removes the item with the specified key name:
d3 = {'name': 'A','year': 2024,'Pass': True,'Marks': [50, 60, 70],'phone_no': 94566,'email': 'asdf@gmail.com'}
del d3['Marks']
```

```python
d3
```

```python
#The del keyword can also delete the dictionary completely:
del d3
```

***Clear Method***

```python
#The clear() method empties the dictionary:
d3 = {'name': 'A','year': 2024,'Pass': True,'Marks': [50, 60, 70],'phone_no': 94566,'email': 'asdf@gmail.com'}
d3.clear()
```

```python
d3
```

***Nested Dictionary***

```python
# Create Nasted Dictionary.
myfamily = {
  "child1" : {
    "name" : "Emil",
    "year" : 2004
  },
  "child2" : {
    "name" : "Tobias",
    "year" : 2007
  },
  "child3" : {
    "name" : "Linus",
    "year" : 2011
  }
}
```

```python
# Print Nasted dictionary.
myfamily
```
