## List item Addition

```python
# List creation
course=["Python","SQL","Stats"]
print(course)
```

## Append Function.

```python
# Use append function for Add new values in end of the list.
course=["Python","SQL","Stats"]
course.append("Excel")
print(course)
```

## Insert Function.

```python
# Use Insert Function for insert new values in list.
course=["Python","SQL","Stats"]
course.insert(3,'Algo')
print(course)
```

## Extend Function

```python
# Use Extend Function for add two lists in one.
course=["Python","SQL","Stats"]
company=['TCS','Infosys','Wipro']
course.extend(company)
print(course)
```

## Remove Method

```python
# The remove() method removes the specified item.
course=['Python', 'SQL', 'Stats', 'Algo']
course.remove('Algo')
print(course)
```

## Pop Method

```python
#If do not specify the index, the pop() method removes the last item.
stud=['M','N','O','P','Q','R']
stud.pop()
```

```python
stud
```

```python
#pop() method removes the specified index.
stud.pop(1)
```

```python
stud
```
