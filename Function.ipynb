```python
# Wap to chek weather a given number is prime number or not
num = int(input('Enter a Numbre'))
if num == 0 or num == 1:
    print(num, 'is not a prime number')
elif num > 1:
    # Check for factors
    for i in range(2,num):
        if (num %i) == 0:
            print(num, 'is not a prime Number')
            break
        else :
            print(num, 'is a prime Number')
# If input number is less then 
# or equal to 1, it is not prime
else :
   print(num, 'is not a prime Number')
```

```python
# Write a function to swap two numbers without usion third varable
def swap(a,b):
    a,b=b,a
    print(a)
    print(b)
```

```python
# Write a function to swap two numbers usion third varable
def swap(a,b):
    temp=a
    a=b
    b=temp
    print(a)
    print(b)
```

```python
swap(457,54)
```
